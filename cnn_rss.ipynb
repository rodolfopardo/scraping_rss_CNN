```python
#Columna de datos con todas las noticias de la CNN 

noticias_df = pd.DataFrame(cnn_rss['entries'])
noticias_df.head(5)
```

```python
#Queremos averiguar cuales son los autores con mas artículos aportados al canal de RSS de CNN 

autores = noticias_df.groupby('author', as_index=False).agg({'title':'count'})
autores
#Ordena los autores
autores.sort_values('title', ascending = False)
```

```python
#Extrayendo los titulos que solo hablan del Chapo Guzmán 

titulos_chapo = []

for x in noticias_df['title']:
    if 'El Chapo' in x:
        titulos_chapo.append(x)
    else:
        pass

print(titulos_chapo)
```
